```python
import pandas as pd
from IPython.display import display

# Cargando el dataframe paises.csv
ruta_paises_csv = 'paises.csv'
df_countries = pd.read_csv(ruta_paises_csv)

# Mostrando las primeras filas del dataframe 'paises.csv'
print("\nPrimeras filas del dataframe 'paises.csv':")
display(df_countries.head())

# Cargando el archivo 'winemag-data-130k-v2.csv' '
ruta_winemag_data = "winemag-data-130k-v2.csv" 
wine_df = pd.read_csv(ruta_winemag_data, index_col=0)

# Mostrando las primeras filas del dataframe 'winemag-data-130k-v2.csv'
print("\nPrimeras filas del dataframe 'winemag-data-130k-v2.csv':")
display(wine_df.head())
```

```python
# Renombrando columnas
wine_df = wine_df.rename(columns={'country': 'pais', 'points': 'puntuacion', 'price': 'precio', 'variety': 'variedad'})

print("\nPrimeras filas del dataframe 'winemag-data-130k-v2.csv' después del renombramiento:")
display(wine_df.head())
```

```python
# Creando nuevas columnas
wine_df['precio_promedio'] = wine_df.groupby('pais')['precio'].transform('mean')
wine_df['vino_caro'] = wine_df['precio'] > wine_df['precio_promedio']
wine_df['rango_puntuacion'] = pd.cut(wine_df['puntuacion'], bins=[0, 80, 90, 100], labels=['Regular', 'Bueno', 'Excelente'])

print("\nPrimeras filas del dataframe 'winemag-data-130k-v2.csv' después de la creación de columnas:")
display(wine_df.head())
```

```python
# Fusionando por país
df_merged = pd.merge(wine_df, df_countries, how='left', left_on='pais', right_on='nombre')

print("\nPrimeras filas del dataframe fusionado:")
display(df_merged.head())
```

```python
# Agrupando por país y continente
df_grouped = df_merged.groupby(['continente', 'pais'])
```

```python
# Reportes por agrupamiento de datos
reporte1 = df_grouped['puntuacion'].mean().sort_values(ascending=False).head(5)
reporte2 = df_grouped['precio'].mean().sort_values(ascending=False).head(5)
reporte3 = df_grouped['variedad'].value_counts().groupby('continente').head(1)
reporte4 = df_grouped['rango_puntuacion'].value_counts().unstack().fillna(0)

# Guardando el reporte3 en un archivo CSV
ruta_reporte3_csv = 'top_variedad_por_continente.csv'
reporte3.to_csv(ruta_reporte3_csv)

print(f"\nEl reporte3 se ha guardado en el archivo CSV: {ruta_reporte3_csv}")
```

```python
# Mostrando reportes
print("\nReporte 1: Puntuación promedio de vinos por país y continente (Top 5):")
display(reporte1)

print("\nReporte 2: Precio promedio de vinos por país y continente (Top 5):")
display(reporte2)

print("\nReporte 3: Top variedad de uva por continente:")
display(reporte3)

print("\nReporte 4: Distribución de rangos de puntuación por continente:")
display(reporte4)
```
