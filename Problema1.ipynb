```python
import pandas as pd

# Cargando el dataset
df_airbnb = pd.read_csv("airbnb.csv")

# CASO 1: Alicia busca alojamiento para su familia

# Filtrando las habitaciones para Alicia
alicia_criteria = (df_airbnb['accommodates'] >= 4) & (df_airbnb['bedrooms'] >= 2) & (df_airbnb['reviews'] > 10) & (df_airbnb['overall_satisfaction'] > 4)

# Ordenando por puntuación y número de críticas
alicia_rooms = df_airbnb[alicia_criteria].sort_values(by=['overall_satisfaction', 'reviews'], ascending=[False, False])

# Tomando las 3 mejores alternativas
top_3_alternatives = alicia_rooms.head(3)

top_3_alternatives
```

```python
# Buscando por id
df_r = df_airbnb.query("room_id == 97503")
df_c = df_airbnb.query("room_id == 90387")

# Uniendo dataframes
df_union = pd.concat([df_r, df_c])
print(df_union)

# Guardando como excel
df_union.to_excel("roberto.xlsx")

print("El archivo Excel roberto.xlsx se creó correctamente")
```

```python
# CASO 3: Diana busca propiedades económicas en Lisboa

# Filtrando las propiedades para Diana
diana_criteria = (df_airbnb['price'] <= 50) & (df_airbnb['room_type'] == 'Shared room')

# Ordenando por precio y puntuación
diana_rooms = df_airbnb[diana_criteria].sort_values(by=['price', 'overall_satisfaction'], ascending=[True, False])

# Tomando las 10 propiedades más baratas
top_10_cheap_properties = diana_rooms.head(10)

top_10_cheap_properties
```
